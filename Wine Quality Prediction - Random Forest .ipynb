Importing the Dependencies 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score
```

Data Collection

```python
# Loading the dataset to a Pandas DataFrame

wine_dataset = pd.read_csv('winequality-red.csv')
```

```python
# Number of rows and columns 

wine_dataset.shape
```

```python
# First 5 rows of the dataset

wine_dataset.head()
```

```python
# Checking for missing values 

wine_dataset.isna().sum()
```

Data Analysis and Visualizations 

```python
# Statistical measures of the dataset

wine_dataset.describe()
```

```python
# Number of values for each quality

sns.catplot( x = 'quality', data = wine_dataset, kind = 'count')
```

```python
corr_matrix = wine_dataset.corr()
```

```python
print(corr_matrix)
```

```python
plt.figure(figsize = (15,10))

sns.heatmap( corr_matrix, annot = True, fmt = ".1f", cmap = 'coolwarm')

plt.title('Correlation Matrix of Wine Quality Dataset')

plt.show()
```

```python
# Alcohol vs Quality 

plt.figure(figsize = (5,5))

sns.boxplot(x='quality', y = 'alcohol', data = wine_dataset)

plt.xlabel('Wine Quality')
plt.ylabel('Alcohol Content')

plt.show()
```

```python
# Volatile Acidity vs Quality 

plot = plt.figure(figsize = (5,5))
sns.barplot(x = 'quality', y = 'volatile acidity', data = wine_dataset)
```

```python
#  Citric Acid vs Quality 

plot = plt.figure(figsize = (5,5))
sns.barplot(x = 'quality', y = 'citric acid', data = wine_dataset)
```

Data Preprocessing 

```python
# Separate the data and Label

x = wine_dataset.drop('quality', axis = 1)

# Label Binarization

y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
x
```

```python
y
```

```python
y.nunique()
```

Train Test Split 

```python
xtrain, xtest, ytrain, ytest = train_test_split( x, y, test_size =0.2, random_state = 3)
```

```python
print(x.shape, xtrain.shape, xtest.shape)
```

```python
print(y.shape, ytrain.shape, ytest.shape)
```

Model Training 

Random Forest Classifier 

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain, ytrain)
```

Model Evaluation

Accuracy Score 

```python
# Accuracy on Test Data 

xtest_pred = model.predict(xtest)

test_data_accuracy = accuracy_score( xtest_pred, ytest )
```

```python
print(test_data_accuracy)
```

Building a Predictive System

```python
input_data = input('Insert the input values')

# Changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshaping the data as we are predicting the label for only 1 instance 
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==1):
    print('Good Quality')
else:
    print('Bad Quality')
```
